```python
import sys

sys.path.insert(1, '../')
from classifier import *
from feature_extraction import *
import matplotlib.pyplot as plt
```

```python
data = pd.read_parquet('/media3/etienne/workdir/spark_AGN/data/allAGN_100000.parquet').sample(frac=0.01, random_state=42)
```

```python
clean = fe.clean_data(data)
converted = fe.convert_full_dataset(clean)
transformed_1, transformed_2 = fe.transform_data(converted)

features_1 = fe.parametrise(transformed_1, k.MINIMUM_POINTS, 1)
features_2 = fe.parametrise(transformed_2, k.MINIMUM_POINTS, 2)
```

```python
obj = 'ZTF18acplnnx'


features_2_bis = features_2.drop(columns={"object_id"})

features = features_1.join(features_2_bis)
valid = features["valid_1"] & features["valid_2"]
features = features.drop(columns=["valid_1", "valid_2"])
features_obj = features[features['object_id'] == obj]

ps = features_obj.iloc[0]

# Compute fitted values at cjd from the other band
add_from_2 = mod.bump(ps["cjd_2"], *ps["bump_1"])
add_from_1 = mod.bump(ps["cjd_1"], *ps["bump_2"])

# Add to the flux list : maintain the same order : cjd from 1 then cjd from 2
new_cflux_1 = np.append(ps["cflux_1"], add_from_2)
new_cflux_2 = np.append(add_from_1, ps["cflux_2"])

new_cjd = np.append(ps["cjd_1"], ps["cjd_2"])

unnorm_cflux_1 = add_from_2 * ps["peak_1"]
unnorm_cflux_2 = add_from_1 * ps["peak_2"]
```

```python
flux1 = ps["cflux_1"]* ps["peak_1"]
flux2 = ps["cflux_2"]* ps["peak_2"]
jd1 = ps["cjd_1"]
jd2 = ps["cjd_2"]

x_high = np.linspace(15, 45, 100)

title_size = 20
figure_size = (12, 7)
xylabel_size = 16
legend_size = 14
marker_size = 100
line_size = 7
axis_size = 2
tick_size = 14

fig, ax1 = plt.subplots(1, 1, figsize=figure_size)

ax1.set_title('Normal spine and ticks')

# change all spines
for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(axis_size)

# increase tick width
ax1.tick_params(width=axis_size)


plt.title("Color computation using the Bump function", fontsize = title_size,fontweight="bold")
plt.xlabel('Time (days)', fontsize = xylabel_size,fontweight="bold")
plt.ylabel('Flux', fontsize = xylabel_size,fontweight="bold")

plt.scatter(jd1, flux1, marker= 'o', color = 'darkolivegreen', label='g band data', s=marker_size, zorder=2)
plt.plot(x_high,  mod.bump(x_high, *ps["bump_1"])* ps["peak_1"], color = 'darkolivegreen', alpha = 0.2, label='g band fit',linewidth=line_size, zorder=1)
plt.scatter(jd2, unnorm_cflux_1, marker= 'x', color = 'darkolivegreen', label='Computed g band data', s=marker_size, zorder=2)


plt.scatter(jd2, flux2, marker= 'o', color = 'firebrick', label='r band data', s=marker_size, zorder=2)
plt.plot(x_high,  mod.bump(x_high, *ps["bump_2"])* ps["peak_2"], color = 'firebrick', alpha = 0.2, label='r band fit',linewidth=line_size, zorder=1)
plt.scatter(jd1, unnorm_cflux_2, marker= 'x', color = 'firebrick', label='Computed r band data', s=marker_size, zorder=2)

plt.yticks(fontsize=tick_size)
plt.xticks(fontsize=tick_size)


plt.legend(fontsize=legend_size)
plt.savefig("EAS_color.png",bbox_inches='tight')
```
